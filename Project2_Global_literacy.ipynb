```python
#standardise the country names

dfs = [df_literacy, df_illiteracy, df_gdp_schooling]

for df in dfs:
    df['entity'] = (
        df['entity']
        .str.strip()          # remove leading/trailing spaces
        .str.title()          # convert to normal case
    )
```

```python
dfs = [df_literacy, df_illiteracy]

all_missing = pd.concat(
    [df.loc[df['code'].isna(), 'entity'] for df in dfs]
).unique()

print(sorted(all_missing))
```

```python
###########droping the regions names since the analysis will be done for availablle country level###
dfs = [df_literacy, df_illiteracy]

for df in dfs:
    df.dropna(subset=['code'], inplace=True)
    
#forward filling in literacy_rate, mf_youth_and_adults__15_64_years__average_years_of_education column--- fills the previous and next years value if missing using backfward and forward filling
df_gdp_schooling = df_gdp_schooling.sort_values(['entity', 'year'])

df_gdp_schooling['literacy_rate'] = (
    df_gdp_schooling.groupby('entity')['literacy_rate']
    .ffill()
    .bfill()
)
df_gdp_schooling['mf_youth_and_adults__15_64_years__average_years_of_education'] = (
    df_gdp_schooling.groupby('entity')['mf_youth_and_adults__15_64_years__average_years_of_education']
    .ffill()
    .bfill()
)
```

```python
df_literacy = df_literacy.dropna()
```

```python
#column names change
dfs = [df_literacy, df_illiteracy, df_gdp_schooling]
for df in dfs:
    df.rename(columns={'entity':'Country'}, inplace=True)
```

```python
#columns addition
df_literacy['all_youth_lit_rate'] = df_literacy['youth_lit_rate_Male'] + df_literacy['youth_lit_rate_Female']
df_literacy['youth_vs_adult_%'] = (df_literacy['all_youth_lit_rate'] / df_literacy['adult_lit_rate']) * 100
df_literacy['lit_gap_M&F'] = df_literacy['youth_lit_rate_Male'] - df_literacy['youth_lit_rate_Female']
df_literacy['youth_lit_avg'] = df_literacy['all_youth_lit_rate'].mean()
df_illiteracy['literacy_growth_%'] = df_illiteracy.groupby('Country')['literacy_rate'].pct_change() * 100
df_illiteracy['literacy_growth_%'] = df_illiteracy['literacy_growth_%'].fillna(0)
df_gdp_schooling['gap'] = 100 - df_gdp_schooling['literacy_rate']
df_gdp_schooling['gdp_per_schooling_year'] = (df_gdp_schooling['ny_gdp_pcap_pp_kd'] / df_gdp_schooling['avg_years_of_edu'])
df_gdp_schooling['gdp_%'] = (df_gdp_schooling['ny_gdp_pcap_pp_kd'] / df_gdp_schooling['ny_gdp_pcap_pp_kd'].sum()) * 100
df_gdp_schooling['education_index'] = df_gdp_schooling['avg_years_of_edu'] / 15
```

```python
#Get top 5 countries with highest adult literacy in 2020.

query1 = """
SELECT Country, year, adult_lit_rate 
FROM literacy_rates
WHERE year = 2020
ORDER BY adult_lit_rate DESC 
LIMIT 5;
"""
cursor.execute(query1)
query1_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query1_result)
```

```python
#Find countries where female youth literacy < 80%

query2 = """
SELECT DISTINCT Country, youth_lit_rate_Female FROM literacy_rates
WHERE youth_lit_rate_Female < 80;
"""
cursor.execute(query2)
query2_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query2_result)
```

```python
#Average adult literacy per continent (owid region)

query3 = """
SELECT owid_region,
AVG(adult_lit_rate) AS Avg_adult_literacy
FROM literacy_rates
GROUP BY owid_region
ORDER BY avg_adult_literacy DESC;
"""
cursor.execute(query3)
query3_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query3_result)
```

```python
#Countries with illiteracy % > 20% in 2000

query4 = """
SELECT Country, year,
illiteracy_rate 
FROM illiteracy_population
WHERE illiteracy_rate > 20 AND year = 2000
ORDER BY Country;
"""
cursor.execute(query4)
query4_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query4_result)
```

```python
# Trend of illiteracy % for India (2000–2020)

query5 = """
SELECT Country, year,
illiteracy_rate 
FROM illiteracy_population
WHERE Country = 'India' AND year BETWEEN '2000' AND '2020'
ORDER BY year;
"""
cursor.execute(query5)
query5_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query5_result)
```

```python
#Top 10 countries with largest illiterate population in the last year.

query6 = """
SELECT t1.Country, t1.year, t1.illiteracy_rate
FROM illiteracy_population t1
JOIN (
    SELECT Country, MAX(year) AS latest_year
    FROM illiteracy_population
    GROUP BY Country
) t2
ON t1.Country = t2.Country 
AND t1.year = t2.latest_year
ORDER BY t1.illiteracy_rate DESC
LIMIT 10;
"""
cursor.execute(query6)
query6_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query6_result)
```

```python
#Find countries with avg_years_schooling > 7 and gdp_per_capita < 5000
query7= """
SELECT Country, year, avg_years_of_edu, ny_gdp_pcap_pp_kd
FROM gdp_schooling
WHERE ny_gdp_pcap_pp_kd < 5000 AND avg_years_of_edu > 7
ORDER BY Country;
"""
cursor.execute(query7)
query7_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query7_result)
```

```python
#Rank countries by GDP per schooling for the year 2020
query8= """
SELECT Country, year, gdp_per_schooling_year
FROM gdp_schooling
WHERE year = 2020
ORDER BY gdp_per_schooling_year
"""
cursor.execute(query8)
query8_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query8_result)
```

```python
#Find global average schooling years per year

query9= """
SELECT 
    year,
    AVG(avg_years_of_edu) AS global_avg_schooling
FROM gdp_schooling
GROUP BY year
ORDER BY year; 
"""
cursor.execute(query9)
query9_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query9_result)
```

```python
#List top 10 countries in 2020 with highest GDP per capita but lowest average years of schooling(less than 6).

query10 = """
SELECT Country, year,
ny_gdp_pcap_pp_kd AS GDP_per_capita,
avg_years_of_edu
FROM gdp_schooling
WHERE year = 2020 AND avg_years_of_edu < 6
ORDER BY ny_gdp_pcap_pp_kd DESC
LIMIT 10;
"""
cursor.execute(query10)
query10_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query10_result)
```

```python
#Show countries where the illiterate population is high despite having more than 10 average years of schooling.

query11 = """
SELECT DISTINCT t1.Country, t1.year,
t1.illiteracy_rate,
t2.avg_years_of_edu
FROM illiteracy_population t1
INNER JOIN gdp_schooling t2
ON t1.Country = t2.Country
AND t1.year = t2.year
WHERE t2.avg_years_of_edu > 10
AND t1.illiteracy_rate > 13
ORDER BY t1.illiteracy_rate, t2.avg_years_of_edu DESC;
"""
#global threshold for illiteracy rate is 13%
cursor.execute(query11)
query11_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query11_result)
```

```python
#Compare literacy rates and GDP per capita growth for a selected country over the last 20 years. (country of your choice)
query12 = """
SELECT t1.Country, t1.year, t1.literacy_rate, t2.ny_gdp_pcap_pp_kd
FROM illiteracy_population t1
INNER JOIN gdp_schooling t2
ON t1.Country = t2.Country
AND t1.year = t2.year
WHERE t1.Country = "India"
AND t1.year BETWEEN 2000 AND 2023
ORDER BY t1.literacy_rate
"""
cursor.execute(query12)
query12_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query12_result)
```

```python
#Show the difference between youth literacy male and female rates for countries with GDP per capita above $30,000 in 2020.

query13 = """
SELECT t1.Country, t1.year, t1.youth_lit_rate_Male, t1.youth_lit_rate_Female,
`t1`.`lit_gap_M&F`,
t2.ny_gdp_pcap_pp_kd
FROM literacy_rates t1
INNER JOIN gdp_schooling t2
ON t1.Country = t2.Country
AND t1.year = t2.year
WHERE t2.ny_gdp_pcap_pp_kd > 30000 AND t1.year = 2020
ORDER BY t1.Country;
"""
cursor.execute(query13)
query13_result = pd.DataFrame(cursor.fetchall(), columns=[c[0] for c in cursor.description])
display(query13_result)
```
